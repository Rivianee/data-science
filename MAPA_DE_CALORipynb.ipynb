```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Criação de um DataFrame de exemplo
data = np.random.rand(10, 10)  # Dados aleatórios
df = pd.DataFrame(data, columns=[f'Feature_{i}' for i in range(10)])

# Criação do clustermap
clustermap = sns.clustermap(df, cmap='viridis')

# Mostra o gráfico
plt.show()
```
